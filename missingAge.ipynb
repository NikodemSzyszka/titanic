```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import os
import string
```

```python
TITANIC_PATH = os.path.join("titanic")
def load_titanic_data(filename, titanic_path=TITANIC_PATH):
    csv_path = os.path.join(titanic_path, filename)
    return pd.read_csv(csv_path)
```

```python
train_data = load_titanic_data("train.csv")
test_data = load_titanic_data("test.csv")
```

```python
data = pd.concat([train_data, test_data])
print(data.isnull().sum())
data.head()
```

```python
#brakuje dużo danych w kategori wiek, wykorzystamy zawarte w Name tytuły żeby lepiej przybliżyć wiek 
data['Title'] = data.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
print(data['Title'].value_counts())
data.Title.unique()
```

```python
titles = ['Don', 'Mme', 'Ms',
       'Major', 'Lady', 'Sir', 'Mlle', 'Col', 'Capt', 'Countess',
       'Jonkheer', 'Dona']
for title in titles:
    display(data[data['Title']==title])
```

```python
#unikalne tytyły wszytskie mają wiek znany poza id 980, dlatego najpierw uzupełniam brakujący wiek potem nadaje tytuły
for title in ['Mr', 'Mrs', 'Miss', 'Master', 'Rev', 'Dr']:
    data.loc[(data.Age.isnull()) & (data.Title==title), 'Age'] = data.groupby('Title').Age.median()[title]

data.loc[(data.Age.isnull()) & (data.Title=='Ms'), 'Age'] = data.groupby('Title').Age.median()[title]
```

```python
data['Title'] = data['Title'].replace('Don','Mr')
data['Title'] = data['Title'].replace('Major','Mr')
data['Title'] = data['Title'].replace('Sir','Mr')
data['Title'] = data['Title'].replace('Col','Mr')
data['Title'] = data['Title'].replace('Capt','Mr')
data['Title'] = data['Title'].replace('Jonkheer','Mr')

data['Title'] = data['Title'].replace('Countess','Mrs')
data['Title'] = data['Title'].replace('Lady','Mrs')
data['Title'] = data['Title'].replace('Dona','Mrs')

data['Title'] = data['Title'].replace('Mme','Miss')
data['Title'] = data['Title'].replace('Ms','Miss')
data['Title'] = data['Title'].replace('Mlle','Miss')
```
